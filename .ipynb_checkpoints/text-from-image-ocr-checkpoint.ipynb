# 1. **Imports:**

```python
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Input, Dense, Activation, LeakyReLU
from tensorflow.keras.layers import Reshape, Lambda, BatchNormalization
from tensorflow.keras.layers import add, concatenate
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import CuDNNGRU
from tensorflow.keras.optimizers import Adadelta
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import ModelCheckpoint
import tensorflow.keras.callbacks
import numpy as np
import cv2
import itertools
from PIL import Image
import matplotlib.pyplot as plt
import random
import string
import os
from collections import Counter
from sklearn.model_selection import train_test_split

print('imported')
```

# 2. **Preparing data:**

### 2.1.  **Parameters:**

```python
# paths to data
train_path = "data/train/"
test_path  = "data/test/"

# possible characters we need to read
letters = [letter for letter in '0123456789ABCEHKMOPTXY']
num_classes = len(letters) + 1

train_files_list = os.listdir(train_path)
test_files_list  = os.listdir(test_path)

rand_img = cv2.imread(train_path + np.random.choice(train_files_list), cv2.IMREAD_GRAYSCALE)
height, width = rand_img.shape[0], rand_img.shape[1]

batch_size = 32

# random seed
np.random.seed(41)
```

### 2.2. **ImageGenerator:**

```python
# transforms text into array of labels
def text_to_labels(text):
    return list(map(lambda x: letters.index(x), text))

# transforms array of labels into text
def labels_to_text(labels):
    return ''.join(list(map(lambda x: letters[x], labels)))

# applies all changes to image
def preprocess_img(filename):
    rand_img = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    rand_img = rand_img / 255 * 2 - 1
    rand_img = rand_img.T
    rand_img = np.expand_dims(rand_img, axis=-1)
    return rand_img

# transforms the output of the nn into a string
def decode_label(out):
    # out : (1, 32, 42)
    out_best = list(np.argmax(out[0, 2:], axis=1))  # get max index -> len = 32
    out_best = [k for k, g in itertools.groupby(out_best)]  # remove overlap value
    outstr = ''
    for i in out_best:
        if i < len(letters):
            outstr += letters[i]
    return outstr

# Generator:
class TextImageGenerator:
    def __init__(self, directory, width, height, batch_size):
        self.batch_size = batch_size
        self.directory = directory
        self.width = width
        self.height = height
        self.filenames = os.listdir(directory)
        self.size = len(self.filenames)
        print(self.size, " images detected...")
     
    def next_batch(self):
        # label length and input length are fixed (licence plates are always same size)
        label_length = np.full(self.batch_size, 8)
        input_length = np.full(self.batch_size, self.width // 2**3 - 2)
        while True:
            x_batch = []
            y_batch = []
            for i in range(self.batch_size):
                file = np.random.choice(self.filenames) # random img from dataset
                img  = cv2.imread(self.directory + file, cv2.IMREAD_GRAYSCALE) # read img in greyscale
                img  = img.T
                img  = img.astype("float32") / 255 * 2 - 1 # normalization
                x_batch.append(img)
                y_batch.append(text_to_labels(file.split('.')[0]))
            
            x_batch = np.expand_dims(np.array(x_batch), axis=-1)
            y_batch = np.array(y_batch)
            inputs = {
                'input': x_batch,
                'label': y_batch,
                'input_length': input_length,
                'label_length': label_length
            }
            output = {'ctc': np.zeros(self.batch_size)}
            yield (inputs, output)
        
# loss function provided by keras
def ctc_lambda_func(args):
    y_pred, labels, input_length, label_length = args
    # the 2 is critical here since the first couple outputs of the RNN
    # tend to be garbage:
    y_pred = y_pred[:, 2:, :]
    return K.ctc_batch_cost(labels, y_pred, input_length, label_length)
```

# 3. **Creating model:**

### 3.1. **Defining architecture:**

```python
# Input
inputs = Input(name='input', shape=(width, height, 1), dtype='float32')

# Convolution layer
X = Conv2D(64, (3, 3), padding='same', kernel_initializer='he_normal')(inputs) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)
X = MaxPooling2D((2, 2))(X)

X = Conv2D(128, (3, 3), padding='same', kernel_initializer='he_normal')(X) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)

X = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(X) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)
X = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(X) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)
X = MaxPooling2D((2, 2))(X)

X = Conv2D(256, (3, 3), padding='same', kernel_initializer='he_normal')(X) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)
X = Conv2D(512, (3, 3), padding='same', kernel_initializer='he_normal')(X) 
X = BatchNormalization()(X)
X = LeakyReLU(alpha=0.3)(X)
X = MaxPooling2D((2, 2))(X)


# CNN to RNN
X = Reshape(target_shape=((19, 2048)), name='reshape')(X)
X = Dense(64, activation='relu', kernel_initializer='he_normal')(X)

# RNN layer
gru_1 = CuDNNGRU(256, return_sequences=True, kernel_initializer='he_normal', name='gru1')(X)
gru_1b = CuDNNGRU(256, return_sequences=True, go_backwards=True, 
                  kernel_initializer='he_normal', name='gru1_b')(X)
reversed_gru_1b = Lambda(lambda inputTensor: K.reverse(inputTensor, axes=1))(gru_1b)

gru1_merged = add([gru_1, reversed_gru_1b])
gru1_merged = BatchNormalization()(gru1_merged)
    
gru_2 = CuDNNGRU(256, return_sequences=True, kernel_initializer='he_normal', name='gru2')(gru1_merged)
gru_2b = CuDNNGRU(256, return_sequences=True, go_backwards=True, 
                  kernel_initializer='he_normal', name='gru2_b')(gru1_merged)
reversed_gru_2b= Lambda(lambda inputTensor: K.reverse(inputTensor, axes=1))(gru_2b)

gru2_merged = concatenate([gru_2, reversed_gru_2b])  # (None, 32, 1024)
gru2_merged = BatchNormalization()(gru2_merged)

# transforms RNN output to character activations:
y_pred = Dense(num_classes, kernel_initializer='he_normal', activation="softmax")(gru2_merged)

label = Input(name='label', shape=[8], dtype='float32')
input_length = Input(name='input_length', shape=[1])    
label_length = Input(name='label_length', shape=[1])

# Keras doesn't currently support loss funcs with extra parameters
# so CTC loss is implemented in a lambda layer
loss_out = Lambda(ctc_lambda_func, output_shape=(1,), name='ctc')([y_pred, label, input_length, label_length])

model = Model(inputs=[inputs, label, input_length, label_length], outputs=loss_out)
model.summary()
```

### 3.2. **Training:**

```python
train_gen = TextImageGenerator(train_path, width, height, batch_size)
test_gen = TextImageGenerator(test_path, width, height, batch_size)

ada = Adadelta()
model.compile(loss={'ctc': lambda y_true, y_pred: y_pred}, optimizer="adadelta")
auto_saves = ModelCheckpoint(filepath='weights.h5', monitor='val_loss', verbose=1, 
                             save_best_only=True, save_weights_only=True, mode='min', period=1)

model.fit_generator(generator=train_gen.next_batch(),
                    steps_per_epoch=(train_gen.size // batch_size),
                    epochs=10,
                    validation_data=test_gen.next_batch(),
                    validation_steps=(test_gen.size // batch_size),
                    callbacks=[auto_saves])
```

### **(Training for a few more epochs)**

```python
model.fit_generator(generator=train_gen.next_batch(),
                    steps_per_epoch=(train_gen.size // batch_size),
                    epochs=5,
                    validation_data=test_gen.next_batch(),
                    validation_steps=(test_gen.size // batch_size),
                    callbacks=[auto_saves])
```

# 4. **Testing:**

```python
test_files = os.listdir(test_path)
test_images = []
for i in range(0, 300):
    img = preprocess_img(test_path + test_files[i])
    test_images.append(img)
    
test_images = np.array(test_images)

inference_model = Model(inputs, y_pred)
pred = inference_model.predict(test_images)
```

```python
cp = 0
for i in range(0, 300):
    predicted = decode_label(np.expand_dims(pred[i], axis=0))
    if (predicted == test_files[i].split('.')[0]):
        cp += 1
    print("True : ", test_files[i].split('.')[0], "  -> Predicted : ", predicted)
    
print(cp)
```
